```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
land_mines = fetch_ucirepo(id=763) 
  
# data (as pandas dataframes) 
X = land_mines.data.features 
y = land_mines.data.targets 
  
# metadata 
print(land_mines.metadata) 
  
# variable information 
print(land_mines.variables) 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming X and y are your dataframes with features and target
# X contains columns ['V', 'H', 'S'], and y contains the column ['M']

# Map each unique class in `y` to a label for the legend
class_map = {1: 'Class 1', 2: 'Class 2', 3: 'Class 3', 4: 'Class 4', 5: 'Class 5'}
features = ['V', 'H', 'S']
colors = ['blue', 'orange', 'green', 'red', 'purple']  # Assign colors for each class

fig, ax = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Loop through each feature to plot the histogram
for i, feature in enumerate(features):
    ax[i].set_xlabel(f"{feature}")
    ax[i].set_ylabel("Frequency")
    # Plot histograms for each class
    ax[i].hist(
        [X.loc[y['M'] == cls, feature] for cls in sorted(y['M'].unique())],
        bins=15,
        label=[class_map[cls] for cls in sorted(y['M'].unique())],
        color=colors[:len(y['M'].unique())],  # Use only as many colors as there are unique classes
        alpha=0.7,
        edgecolor='black'
    )
    ax[i].legend()
    ax[i].set_title(f"Distribution of {feature} by Class")

fig.tight_layout()
plt.show()
```
